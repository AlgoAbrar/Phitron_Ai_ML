# File Handling

```python
# File Read & Write

# open("Location of the file", "mode of operation") 
# mode of operation:
# "r" - read (default)
# "w" - write (creates a new file or overwrites an existing file)
# "a" - append (creates a new file or appends to an existing file)
# "x" - create (creates a new file, but raises an error if the file already exists) 

# got to goolge collab link

file= open("./sample_data/sample.txt", 'r')
content= file.read()

print(content)
print(type(content))
```

```python
file= open("./sample_data/sample.txt", 'r')
content= file.readlines()

print(content)
print(type(content))

file.close()
print(file.closed)
```

```python
with open("./sample_data/sample.txt", 'r') as file:
    content=file.readlines()
    print(content)
    
print(file.closed)
```

```python
with open("./sample_data/sample.txt", 'r') as file:
    for line in file:
        print(line,end=" ")
```

```python
with open("./sample_data/sample.txt", 'r') as file:
    for line in file:
        l= line.strip()
        print(l)
```
