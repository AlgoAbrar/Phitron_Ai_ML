## Manual Implementation

```python
# prediction f(x)= w1*x1 + w2*x2 + w3*x3 + w4*x4 + b
# for n inputs f(x) = w1*x1 + w2*x2 + w3*x3 + w4*x4 + ... + wn*xn + b
def make_prediction(X,y,W,b):
    m=X.shape[0] # sample number
    pred_list= np.zeros((m,)) # array to store result
    
    for i in range(m):
        # X[i] is the feature vector, W= weights_vector (w is a vector) vector should be capital.
        pred_list[i]= np.dot(W,X[i])+b
    
    return pred_list
```

```python
m=X.shape[0]
n= X.shape[1]

W_init= np.ones((n,))
b=1

make_prediction(X,y,W_init,b)
```

#### Cost Function

```python
# (E1+E2+...+Em)/2m = Error function mean square error
def compute_cost(X,y,W,b):
    m = X.shape[0]
    cost= 0.0
    pred_list= make_prediction(X,y,W,b)
    
    error=pred_list-y # vector of error
    error_squared = error ** 2
    cost = np.sum(error_squared)
    
    cost= cost/(2*m)
    return cost
```

```python
W_init = np.ones((n,))
b_init = 1.0
compute_cost(X,y, W_init ,b_init)
```
