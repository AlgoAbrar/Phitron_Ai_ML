# Pandas For Machine Learning

```python
# Data
import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve", "Frank", "Grace", "Hannah","Sakib"],
    "City": ["New York", "Los Angeles", "Newark", "Boston", "New Delhi", "Chicago", "New Orleans", "Houston","H Los Ang"],
    "Department": ["HR", "IT", "Finance", "IT", "HR", "Marketing", "Finance", "HR", "HR"],
    "Salary": [50000, 60000, 55000, 70000, 52000, 58000, 62000, 51000,70000]
}

df = pd.DataFrame(data)
```

## Filtering on String

```python
df
```

```python
df.loc[df['City'].str.contains('New')]
```

```python
df.loc[df['City'].str.contains('new', case=False)]
```

```python
# Starts with Los
df.loc[df['City'].str.contains(r'^los', case=False)]
# r'^los' starts with los
# r'los' contains los
```

```python
#end with rk
df.loc[df['City'].str.contains(r'rk$', case=False)]
```

```python
# Name starts with vowel

df.loc[df['Name'].str.contains(r'^[AEIOU]', case=False)]
```

```python
# City has new and los

df.loc[df['City'].str.contains(r'new|los', case=False)]
```
