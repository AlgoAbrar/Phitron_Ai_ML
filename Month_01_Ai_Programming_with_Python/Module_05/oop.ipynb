# Object Oriented Programming with Python

### Attributes are variable name, Method is function() name

Class and Object

example: Phone
There can be multiple phone brands, type, battery, processor, camera etc.
Everything has some common structure

Object is call instance in other language

```python
# base class / parent class
class Phone:
    category = "Electronics"
    
    # constructor
    def __init__(self, model, battery, camera):
        self.model=model
        self.battery=battery
        self.camera=camera
        pass
```

```python
# Creating object

apple= Phone()
blueberry = Phone()
motorola = Phone()

print(type(apple))
print(apple.category)
print(blueberry.category)
apple.category= "Super Electronics"
print(apple.category)

# no values
```

```python
# Attribute dite hobe
apple= Phone("Iphone 17", 3000, 40)
blueberry = Phone("B-17", 4000, 12)
motorola = Phone("G50", 3500, 33)
```

```python
print(apple.camera)
print(motorola.camera)
print(motorola.battery)
```

# Inheritance

```python
# Child class / Derived Class

class SmartPhone(Phone):
    def __init__(self, model, battery, camera, processor):
        super().__init__( model, battery, camera)
        self.processor = processor
        
    def charge(self, hour):
        print("Fast charging in process")
        super().charge(hour)
```

```python
pro = SmartPhone("x", 5000,100, "Snapdragon")
print(pro.processor)
pro.charge(1)
```

## Multiple Inheritance

```python
class Cooling_Mechanism:
    def __init__(self, cooling_method):
        self.cooling_method= cooling_method
        
    #methods
    def cooling_on(self):
        print(f"the system is being cool by {self.cooling_method}")
```

```python
class SmartPhone_CoolingMode(SmartPhone,Cooling_Mechanism):
    def __init__(self, model, battery, camera, processor, cooling_method):
        SmartPhone.__init__(self, model, battery, camera, processor) # jehetu multiple korci
        Cooling_Mechanism.__init__(self, cooling_method)
```

```python
pro_cooling = SmartPhone_CoolingMode("Y", 8000, 100, "SnapX", "Nitrogen")

print(pro_cooling.processor) # SmartPhone Class theke
print(pro_cooling.cooling_method) # Phone Class theke
print(pro_cooling.charge(1)) # Cooling_Mechanism class
print(pro_cooling.battery) # Cooling_Mechanism class
print(pro_cooling.cooling_on()) # SmartPhone er modified charge Jeta Phone Class theke inherit
```
