## Manual Implementation

```python
# prediction f(x)= w1*x1 + w2*x2 + w3*x3 + w4*x4 + b
# for n inputs f(x) = w1*x1 + w2*x2 + w3*x3 + w4*x4 + ... + wn*xn + b
def make_prediction(X,y,W,b):
    m=X.shape[0] # sample number
    pred_list= np.zeros((m,)) # array to store result
    
    for i in range(m):
        # X[i] is the feature vector, W= weights_vector (w is a vector) vector should be capital.
        pred_list[i]= np.dot(W,X[i])+b
    
    return pred_list
```

```python
m=X.shape[0]
n= X.shape[1]

W_init= np.ones((n,))
b=1

make_prediction(X,y,W_init,b)
```

#### Cost Function

```python
# (E1+E2+...+Em)/2m = Error function mean square error
def compute_cost(X,y,W,b):
    m = X.shape[0]
    cost= 0.0
    pred_list= make_prediction(X,y,W,b)
    
    error=pred_list-y # vector of error
    error_squared = error ** 2
    cost = np.sum(error_squared)
    
    cost= cost/(2*m)
    return cost
```

```python
W_init = np.ones((n,))
b_init = 1.0
compute_cost(X,y, W_init ,b_init)
```

## Gradient Descent

```python
'''
cost minimizetion using gradient descent
dj is cost function derivative
w= w_old - (alpha*(dj=dw))
b= b_old - (alpha*(dj=db))
update till cost is minimum and convergence is achieved

(dj/dw1)= (1/m) * sum((f(xi)-yi)*xi1)
sum of prediction error * feature1 value for all sample
'''
```

```python
def calculate_gradient(X, y, W, b):
  """Computes the gradient (direction to move) for w and b"""
  m = X.shape[0]
  n = X.shape[1]

  dj_dw = np.zeros((n,)) ## jotota feature totota weights
  dj_db = 0.0 # Change needed for bias

  for i in range(m):
    prediction = np.dot(W,X[i]) + b
    error = prediction - y[i]
    dj_db = dj_db +  error

    for j in range(n):
      dj_dw[j] = dj_dw[j] + (error * X[i,j])



  return dj_dw / m, dj_db / m
     
```

```python
W_init = np.ones((n,))
b_init = 1.0
calculate_gradient(X,y, W_init ,b_init)
```

```python
def gradient_descent(X, y, w_input, b_input, max_iter, alpha=0.01):
  """Automates the tuning of w and b to minimize cost"""
  w = w_input
  b = b_input
  cost_memo = []
  iteration = []

  for i in range(max_iter):
    # 1. Calculate the gradients (the slope)
    dj_dw, dj_db = calculate_gradient(X, y, w, b)

    # 2. Update parameters by taking a small step (alpha) against the gradient
    w = w - alpha * dj_dw
    b = b - alpha * dj_db


    # 3. Track progress by saving the cost
    cost = compute_cost(X,y, w, b)
    cost_memo.append(cost)
    iteration.append(i)

    # Log the status every 100 steps
    if i % 100 == 0:
      print(f"Iteration {i}: Cost {cost:0.4f},dj_dw: {dj_dw} ,dj_db : {dj_db} w: {w}, b: {b:0.4f}")

  return w, b, cost_memo, iteration
```

### Training the Model

```python
W_init = np.zeros((n,))
b_init = 0.0

w_final, b_final, cost_memo, iter_list = gradient_descent(X, y, w_input=W_init, b_input=b_init, max_iter=100000, alpha=0.01)
```

```python
print(w_final)
print(b_final)

# w =[5,10,-2,0.5] , b=2.0
```
