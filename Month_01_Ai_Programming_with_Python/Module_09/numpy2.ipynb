## nd-array manipulation

```python
import numpy as np

arr = np.random.randint(1,100,size=(10,5)) # sample -10 , feature -5
arr.shape
```

```python
# shape change (example change it like 10 row will be 5 col and 5 col will be 10 row)

b=arr.reshape(5,10)

print(b.shape)

# shape should be valid 10*5=50 data can not be converted into 3*5=15 data

print(b.ndim)
print(b)
```

```python
flatten = b.flatten()
print(flatten.ndim)
print(flatten)
```

```python
column_wise_flattening= np.ravel(b,order="F")
print(column_wise_flattening)
```

```python
a= np.random.randint(1,10,size=(2,3))

b= np.random.randint(20,30, size = (2,3))

print(a)
print(b)

```

```python
# Concatenate (row wise or column wise) shape same thakbe

con_row= np.concatenate((a,b),axis=0) # row wise column same thakbe hobe

print(con_row)
```

```python
# Concatenate (row wise or column wise)

con_col= np.concatenate((a,b),axis=1) # col wise row same thakbe hobe

print(con_col)
```

```python
# error

# row wise concetanate korar shomoi column same thakbe
# col wise concat korar shomoi row ta same thakte hobe

x = np.random.randint(1 ,10 , size =(2,3))
y = np.random.randint(1 ,10 , size =(1,3))

r = np.concatenate((x,y),axis =0 )
r = np.concatenate((x,y),axis =1 )
print(r)
```

```python
# transpose

mat = np.array([[10,20,30],
                [30,40,50]
               ])


transpose = mat.T

print(transpose)
```

```python
# array split

a = np.random.randint(1 ,10 , size =(10,))
print(a)
splitted_array = np.array_split(a,3)

print(splitted_array)

# split -> equal division
splitted_array = np.split(a,3) #---> error
```
