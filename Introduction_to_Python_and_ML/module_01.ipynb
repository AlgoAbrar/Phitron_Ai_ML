## Operators in Python

```python
x = 10 ; y = 7 

sum = x + y # sum
sub = x - y # sub
mult = x * y # mult
div = round(x / y , 2) # div

rem = x % y # remainder 

print(sum , sub , mult, div , rem)
```

```python
x = 10 ; y = 7

greater_than = x > y 
greater_than_equal = (10 >= 10)

print(greater_than , greater_than_equal)

less_than = x < y 
less_than_equal = (10 <= 10)

print(less_than , less_than_equal)

equal =  x==y
print(equal)

not_equal = x != y
print(not_equal)
```
