## nd-array manipulation

```python
import numpy as np

arr = np.random.randint(1,100,size=(10,5)) # sample -10 , feature -5
arr.shape
```

```python
# shape change (example change it like 10 row will be 5 col and 5 col will be 10 row)

b=arr.reshape(5,10)

print(b.shape)

# shape should be valid 10*5=50 data can not be converted into 3*5=15 data

print(b.ndim)
print(b)
```

```python
flatten = b.flatten()
print(flatten.ndim)
print(flatten)
```

```python
column_wise_flattening= np.ravel(b,order="F")
print(column_wise_flattening)
```

```python
a= np.random.randint(1,10,size=(2,3))

b= np.random.randint(20,30, size = (2,3))

print(a)
print(b)

```

```python
# Concatenate (row wise or column wise) shape same thakbe

con_row= np.concatenate((a,b),axis=0) # row wise column same thakbe hobe

print(con_row)
```

```python
# Concatenate (row wise or column wise)

con_col= np.concatenate((a,b),axis=1) # col wise row same thakbe hobe

print(con_col)
```

```python
# error

# row wise concetanate korar shomoi column same thakbe
# col wise concat korar shomoi row ta same thakte hobe

x = np.random.randint(1 ,10 , size =(2,3))
y = np.random.randint(1 ,10 , size =(1,3))

r = np.concatenate((x,y),axis =0 )
r = np.concatenate((x,y),axis =1 )
print(r)
```

```python
# transpose

mat = np.array([[10,20,30],
                [30,40,50]
               ])


transpose = mat.T

print(transpose)
```

```python
# array split

a = np.random.randint(1 ,10 , size =(10,))
print(a)
splitted_array = np.array_split(a,3)

print(splitted_array)

# split -> equal division
splitted_array = np.split(a,3) #---> error
```

## Arithmetical & Mathematical

```python
x = np.array([ 10 , 8 , 30 , 100 ])
y = np.array([ 2 , 3 ,4  ,5 ])

add = np.add(x,y)
print(add)

sub = x-y
print(sub)

mul = x*y
print(mul)

div = x/y
print(div)

remainder = x % y

print(remainder)

# faster for vectorization
```

```python
# mathematical function

# trigonometry

# sin , cos , tan takes values in radian

sin_val = np.sin(x)
print(sin_val)

cos_val = np.cos(x)
print(cos_val)

deg_con = np.rad2deg(x)
print(deg_con)

# deg2rad
```

```python
#log

x = np.array([ 10 , 8 , 16 , 100 ])
base_10_log_val = np.log10(x)
print(base_10_log_val)


two_diye_bhag = np.log2(x)
print(two_diye_bhag)

sqrt= np.sqrt(x)
print(sqrt)
```

```python
# different
s = np.sum(x)
print(s)

#prefix sum
cumul =np.cumulative_sum(x)
print(cumul)
```

## Logical Function

```python
# comparison - >,< , >= ,<= , == , !=
import numpy as np

x = np.array([ 10 , 8 , 16 , 100 ])
y = np.array([ 2 , 3 ,16  ,5 ])

greater_than = x>y

print(greater_than)

equal = x==y

print(equal)

# all () -> shob true kina
# any() -> jekono ekta true ase kina

print(np.all(greater_than))

print(np.any(equal))
```

## Sorting

```python
#inplace
x = np.array([ 10 , 8 , 16 , 100 ])

z = x.copy()

print(z)

z.sort()

print(z)
```

```python
# copy sorting
print(x)
sort_arr = np.sort(x)

print(sort_arr)

print(x)
```

```python
# 2d array sorting
mat = np.array([[10,3,5],[8,4,9]])

# horizon axis = 1
# Vertical axis = 0

hor_sort = np.sort(mat,axis=1)
print(hor_sort)

vert_sort = np.sort(mat,axis=0)
print(vert_sort)
```

## counting in ndarray

```python
a = np.random.randint(1 ,100 , size =(100,))
print(a)
```

```python
value_great_than_60 = np.count_nonzero(a>1)
print(value_great_than_60)
```
