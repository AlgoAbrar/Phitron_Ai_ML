# Numerical Python

```python
import numpy as np

# Ndarray

arr1=np.array([1,2,3,4,5])

print(type(arr1))
print(f"Dimension of array is {arr1.ndim}")

# 2d Array

arr2= np.array([[1,2,3,], # row 1
                [4,5,6]]) # row 2

print(f"Dimension of array is {arr2.ndim}")


# 3d array

arr3= np.array([
               #1st floor
               [ [1,2,3], # row 1
                [4,5,6]] ,  # row 2
               # 2nd floo
               
               [ [1,2,3,], # row 1
                [4,5,6]], # row 2
               
               #3rd floor
               [ [1,2,3,], # row 1
                [4,5,6]]] # row 2 
               )

print(f"Dimension of array is {arr3.ndim}")
```

## Attribute of Array

```python
# dimension
print("Dimensions")
print(arr1.ndim)
print(arr2.ndim)
print(arr3.ndim)

#shape
print("")
print("Shape")
print(arr1.shape)
print(arr2.shape)
print(arr3.shape)

# Data type
print("")
print("Data Type")
print(arr1.dtype) # row 5 col 0
print(arr2.dtype) # row 2 column 3
print(arr3.dtype) # floor 3 row 2 col 3

#Size
print("")
print("Size")
print(arr1.size)
print(arr2.size)
print(arr3.size)

```
