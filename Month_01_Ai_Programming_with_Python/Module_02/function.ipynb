# Positional Argument

```python
# functional e thakle parameter
# Bahire thakle Argument
#Aurgent 2 types. Positional & Keyword

def adder(a,b): 
    print(f"{a}, {b}")
    
adder(120,99) #positional argument
```

```python
def adder(*args): 
    for i in args:
        print(i)
        
adder(10,30,20,40)
```

# keyword argument

```python
#keyword argument

def adder(**kwargs):
    print(kwargs)
    
adder(a=10, b=20, c=40, n_estimate=50)
```

```python
#keyword argument

def adder(**kwargs): # value na dewa thakle default value niye kaj kora jai
    for key,value in kwargs.items():
        print(f"{key}: {value}")
    
adder(a=10, b=20, c=40, n_estimate=50)
```

## Return Type of a Function

```python
def welcome_message():
    print("Hello Abrar")
    
print(type(welcome_message()))
```

```python
def summation(a,b): # 3. Parameter, Return Function
    sum=a+b
    return sum


print(type(summation(23,5)))
```

```python
def summation(a,b): # 3. Parameter, Return Function
    sum=a+b
    return sum


print(type(summation(23,5.5)))
```

```python
def welcome_message():
    return f"hello"
    
print(type(welcome_message()))
```

```python
# we can return multiple value
def summation(a,b): # 3. Parameter, Return Function
    return a,b

print(type(summation(10,56.7)))

# unzip tuple

x,y = summation(10.5,20)

print(x,y)
```

## Generator

```python
# generator

lst = [ x for x in range(500)] # assume its a dataset

def data_loader(chunk_size, lst):
    for i in range(0,len(lst),chunk_size):
        yield lst[i:i+chunk_size]
        
x= data_loader(5,lst)

print(type(x)) #class type generator
print(x)

print(next(x))
print(next(x))
```

# Lambda, Map, Filter and Apply Function

```python
# Example

import pandas as pd

df= pd.DataFrame({'temperature (F)':[32,212,98.6,68]})

df

df['temperature (C)']= df['temperature (F)'].apply(lambda x: (x-32)*5/9)

print(df)
```
