# Function in Python in Lens of ML

```python
# Built-in Function (idea is abstracted, we dont know how it works.)
# User Defined Function (idea is open. We know how it works. Can be modified to perform specif tasks)

""""
def name (parameters/input)
-> doc string
 Instrustion
return"""

def welcome(): # 1. No Parameter, No Return type function
    print("hello")
    
# Function Call na korle kaj korbe na

welcome()
```

```python
def welcome_message(user="None"): #2. Parameter, No Return Type Function
    print(f'hello, {user}')

#name= input()
welcome_message()
```

```python
def summation(a,b): # 3. Parameter, Return Function
    sum=a+b
    return sum

print(summation(23,5))
```

# Positional Argument

```python
# functional e thakle parameter
# Bahire thakle Argument
#Aurgent 2 types. Positional & Keyword

def adder(a,b): 
    print(f"{a}, {b}")
    
adder(120,99) #positional argument
```

```python
def adder(*args): 
    for i in args:
        print(i)
        
adder(10,30,20,40)
```

# keyword argument

```python
#keyword argument

def adder(**kwargs):
    print(kwargs)
    
adder(a=10, b=20, c=40, n_estimate=50)
```

```python
#keyword argument

def adder(**kwargs): # value na dewa thakle default value niye kaj kora jai
    for key,value in kwargs.items():
        print(f"{key}: {value}")
    
adder(a=10, b=20, c=40, n_estimate=50)
```
