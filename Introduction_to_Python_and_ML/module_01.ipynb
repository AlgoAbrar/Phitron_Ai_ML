## Taking Input in Python

```python
name = input("What is your name? ")

print(name)
```

```python
age = input("What is your age? ")

print(age , type(age))
```

```python
age = input("Age ?") # string input niye fele 
age = int(age) # string to int convert

print(age , type(age))
```

```python
height = input("What is your height? ")

print(height , type(height))
```

```python
height = float(input("Height ?")) # string input niye fele, tarpor string to float convert
print(height , type(height))
height = height +0.2

print(height)
```

## Operators in Python

```python
x = 10 ; y = 7 

sum = x + y # sum
sub = x - y # sub
mult = x * y # mult
div = round(x / y , 2) # div

rem = x % y # remainder 

print(sum , sub , mult, div , rem)
```
