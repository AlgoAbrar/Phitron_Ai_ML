```python
# Question 1 Solution using np.array()
import pandas as pd
import numpy as np

data = pd.read_csv("staff_data.csv")

emp_ages = np.array(data["emp_age"])
emp_salary = np.array(data["emp_salary"])

# Emp ages
print("Emp ages:")
print(emp_ages)
print(f"Data Type: {emp_ages.dtype}")
print(f"Dimension: {emp_ages.ndim}")
print(f"Shape: {emp_ages.shape}")
print(f"Size: {emp_ages.size}")

print("")

# Emp salary
print("Emp salary:")
print(emp_salary)
print(f"Data Type: {emp_salary.dtype}")
print(f"Dimension: {emp_salary.ndim}")
print(f"Shape: {emp_salary.shape}")
print(f"Size: {emp_salary.size}")
```

```python
# Answer to the question 2
max_emp_salary = np.max(emp_salary)
print(max_emp_salary)

min_emp_salary = np.min(emp_salary)
print(min_emp_salary)

average_emp_salary = np.mean(emp_salary)
print(average_emp_salary)

avg_emp_age=np.mean(emp_ages)
print(avg_emp_age)
```

```python
new_ages = emp_ages[(emp_ages >= 25) & (emp_ages <= 40) & (emp_salary > 50000)]
new_salary = emp_salary[(emp_ages >= 25) & (emp_ages <= 40) & (emp_salary > 50000)]

print("Ages:")
print(new_ages)
print("")

print("Salaries:")
print(new_salary)
print("")

print("Number of employees:")
print(len(new_salary))
```

```python
avg_emp_age=np.mean(emp_ages)
print(f"average age is {avg_emp_age}")

arr_copy = np.copy(emp_ages)

arr_copy[arr_copy>= avg_emp_age] += 2
arr_copy[arr_copy< avg_emp_age] += 4

print("Original Ages:")
print(emp_ages)
print("")

print("Modified Ages:")
print(arr_copy)
```

```python
roles = np.array(data["role"])

techrole = (roles == "Software Developer") | (roles == "Backend Developer")

emp_salary_float = emp_salary.astype(float)

old_total = np.sum(emp_salary_float[techrole])

new_emp_salary = emp_salary_float.copy()
new_emp_salary[techrole] *= 1.08

new_total = np.sum(new_emp_salary[techrole])

increase = new_total - old_total

print("Old Total Salary:")
print(old_total)
print("")

print("New Total Salary:")
print(new_total)
print("")

print("Increase Expense:")
print(increase)
```

```python
newdf = df.copy()

newdf = newdf.sort_values(
    by=["emp_age", "emp_salary"],
    ascending=[True, False]
)

newdf = newdf[["emp_id", "emp_salary", "role"]]
newdf.head(5)
```

```python
df["previous_salary"] = df["emp_salary"]
df["new_salary"] = df["previous_salary"]

techroles = (df["role"] == "Software Developer") | (df["role"] == "Backend Developer")

non_techroles = (df["role"] != "Software Developer") & (df["role"] != "Backend Developer")


df.loc[techroles, "new_salary"] = df["previous_salary"] * 1.12

df.loc[non_techroles, "new_salary"] = df["previous_salary"] * 1.05

df["increase_of_salary"] = df["new_salary"] - df["previous_salary"]


print("Roles:")
print(df["role"])
print("")

print("Previous Salary:")
print(df["previous_salary"])
print("")

print("New Salary:")
print(df["new_salary"])
print("")

print("Increase Salary:")
print(df["increase_of_salary"])
```
