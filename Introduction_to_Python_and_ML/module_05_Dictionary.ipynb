# Dictionary

```python
# Dictionary keeps key and a value
# It is unordered so no indexing
# Format (key, Value)

dict_a={"Name":"Abrar", "Age":21, "City":"Rajshahi", "marks":[90,95,85]}
print(type(dict_a),dict_a)
```

```python
# Empty Dictionary
dict_b={}
print(type(dict_b),dict_b)

```

```python
# To access the value of a key we can use the key name
print(dict_a["Name"])
print(dict_a["Age"])
```

```python
dict_a={"Name":"Abrar", "Age":21, "City":"Rajshahi", "marks":[90,95,85]}
print(dict_a)

print(dict_a.get("math_marks:"))
print(dict_a.get("math_marks:",0))
```

```python
# Add value
dict_a={"Name":"Abrar", "Age":21, "City":"Rajshahi", "marks":[90,95,85]}
print(dict_a)
dict_a["math_marks"]=30

print(dict_a)

dict_a.update({"English_marks":30})
print(dict_a)
```

```python
#Delete a key
print(dict_a)
del dict_a["math_marks"]
print(dict_a)
```

```python
dic2= dict_a.copy()
print(dic2)

# Key will be immutable but value can be mutable
```

```python
#testing

dic3={[10,20,30]:"a"} # key should be immutable (hashable) but list is mutable (unhashable) so it will give error

print(dic3)
```

```python
#testing

dic3={(10,20,30):"a"} # it will work

print(dic3)
```

```python
dic4={"key": {"Name": "Abrar", "Age": 21,"City":"Rajshahi"}} # value can be mutable

print(dic4)
```
