## Tuple

```python
# Tuple is Collection of data which is immutable
print("Tuple is Collection of data which is immutable")
```

```python
# declaration of tuple

tup = (10,20,30,40,50)

float_tup = (10.5,20.5,30.5,40.5,50.5)

#print(type(float_tup),float_tup)

mixed_tup = (10, 10.5, "Abrar", True)

tup1=tuple([10]) #converting list to tuple
print(type(tup1),tup1)

lst=[10,20,30,40,50]
tup2=tuple(lst) #converting list to tuple
print(type(tup2),tup2)
```

```python
# Accessing tuple elements
tup = (10,20,30,40,50, 10.5, "Hello", True)

print(tup[1])
print(tup[6])
print(tup[7])

new_tup=tup[3:7] #slicing of tuple
print(new_tup)
```

```python
# Immutability of tuple
# If we have a list we can change the value of list but in tuple we cannot change the value of tuple because tuple is immutable
tup[0]=100 # this will give error because tuple is immutable

print(tup)
```

```python
# Tuple Methods
tup = (10,20,30,40,50, 10.5, "Hello", True,10,10,10,10.5,10.5,10.5,10.5)
print(tup.count(10)) # it will count how many times 10 is present in the tuple
print(tup.index(10.5)) # it will give the index of 10.5
```

## Set

```python
# Set stores unique values and it is mutable and unordered collection of data
print("Set stores unique values and it is mutable")

sett={10,20,30,40,50, 10.5, "Hello", True}
print(type(sett),sett)
```

```python
set_b={}
print(type(set_b),set_b) #dict means dictionary and it is empty dictionary not empty set
```

```python
s=set() # this is how we can declare an empty set
print(type(s),s)
```

```python
set_s=(10,20,30,40,50, 10.5, "Hello", True,10,10,10,10.5,10.5,10.5,10.5)

print(set(set_s)) # it will remove the duplicate values from the set
```

```python
if(10 in set(set_s)):
    print("10 is present in the set")
else:
    print("10 is not present in the set")
```

```python
set_s=(10,20,30,40,50)
sum=0
for element in set_s:
    sum+=element
print(sum)
```

```python
S={5,1,2,4,3}

S.add(10    ) # it will add 10 to the set
S.add(0) # it will add 0 to the set
S.add(3) # it will not add 3 to the set because 3 is already present in the set
print(S)

S.pop() # it will remove a random element from the set
print(S)

S.pop()
print(S)

S.remove(4) # it will remove 4 from the set
print(S)
```

```python
sett_A={10,20,30,40,50}
sett_B={40,50,60,70,80}

#Union and Intersection of sets

# Union of sets
uni_set=sett_A.union(sett_B)
print(uni_set)

# Intersection of sets
inter_set=sett_A.intersection(sett_B)
print(inter_set)

# Disjoint sets
dis_set=sett_A.isdisjoint(sett_B)
print(dis_set)

sett_c={90,91,92,93,94}
dis_set=sett_B.isdisjoint(sett_c)
print(dis_set)
```
