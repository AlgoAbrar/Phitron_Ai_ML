```python
def personal_details(name,age):
    print(f"My name is {name} and I am {age} years old.")
    
# input in a single line separated by space
n= input().split()
personal_details(n[0], int(n[1]))
```

# Problem 2

```python
def batch_generator(data, batch_size):
    batches = []
    for i in range(0, len(data), batch_size):
        batches.append(data[i:i + batch_size])
    return batches

batch_size = int(input())
data = list(map(int, input().split()))
for batch in batch_generator(data, batch_size):
    print(batch)
```

# Problem 4

```python
'''  Write a function power(base, exponent) that returns the result of base raised to exponent. If the exponent is not given, it should calculate the square.'''

def powerr(base,exp=2):
    return base**exp
    
data = input().split()
data = list(map(int, data))

if len(data) == 1:
    base = data[0]
    print(powerr(base))
else:
    base = data[0]
    exp = data[1]
    print(powerr(base, exp))
```
