## Conditional Statement

```python
num = int(input("Enter a number: "))

if num > 0 and num % 2 == 0:
    print("Positive even number")

elif num > 0 and num % 2 != 0:
    print("Positive odd number")

elif num < 0:
    print("Negative number")

else:
    print("The number is zero")
```

## Loop

```python
for i in range(1, 21):
    if i % 3 == 0 and i % 5 == 0:
        print(i, "Abrar")
    elif i % 3 == 0:
        print(i, "Rhyme")
    elif i % 5 == 0:
        print(i, "Saiyedul")
    else:
        print(i)
```

## List

```python
numbers = [5, 12, 8, 3, 15]

result = []

for num in numbers:
    if num > 10:
        result.append(num * 2)
    else:
        result.append(num)

print(result)
```

## Dictionary

```python
marks = {
    "Alice": 85,
    "Bob": 45,
    "Charlie": 60,
    "David": 30
}

for name in marks:
    if marks[name] >= 50:
        print(name, "passed")
    else:
        print(name, "failed")
```
