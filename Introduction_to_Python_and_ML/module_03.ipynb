## Tokenize

### splitting joining and formatted string

```python
prompt = "what is your name ?"

token = prompt.split()
print(token)

#join to convert into string again
new_string = " ".join(token)
print(new_string)
print(type(new_string))
```

```python
#formatting string
name = "Abrar"
age=24
height= 6.01123445

#print(f"His name is {name} and his age is {age} and his height is {height:.2f} feet")
info= (f"His name is {name} and his age is {age} and his height is {height:.2f} feet")

print(info)
print(type(info))


model_accuracy= 0.83333
print(f"The model accuracy is {model_accuracy:.2%}")
```
