# Dictionary

```python
# Dictionary keeps key and a value
# It is unordered so no indexing
# Format (key, Value)

dict_a={"Name":"Abrar", "Age":21, "City":"Rajshahi", "marks":[90,95,85]}
print(type(dict_a),dict_a)
```

```python
# Empty Dictionary
dict_b={}
print(type(dict_b),dict_b)

```

```python
# To access the value of a key we can use the key name
print(dict_a["Name"])
print(dict_a["Age"])
```
