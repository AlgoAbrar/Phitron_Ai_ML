# Question 2

```python
def mathfu(a, b):
    try:
        res = ((a*a)+(b*b))/(a-b)
        return res
    except Exception as e:
        return (f"Error: {e}. Division is not possible")

inp= input()
numbers = inp.split()
print(mathfu(a=int(numbers[0]),b=int(numbers[1])))
```

# Question 3

```python
with open("./sample_data/data.txt", 'r') as file:
    content=file.read()
    content=content.lower()
    #print(content)
    words= content.split()
    #print(words)
    freq={}
    
    for word in words:
        if word in freq:
            freq[word] += 1
        else:
            freq[word] = 1

    maxc = 0
    theword = ""

    for word in freq:
        if freq[word] > maxc:
            maxc = freq[word]
            theword = word
    #print(maxc)
    print(theword)
```
