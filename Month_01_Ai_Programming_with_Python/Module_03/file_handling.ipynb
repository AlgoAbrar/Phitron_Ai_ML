# File Pointer

```python
with open("./sample_data/sample.txt","r") as file:
    print(file.tell()) #current cursor
    print(file.read())
    print(file.tell()) # after reading the whole file, the cursor is at the end of the file
    print(file.read()) # since the cursor is at the end of the file, it returns an empty string
    file.seek(0) # move the cursor to the beginning of the file
    print(file.read()) # now it reads the whole file again
```

```python
with open("./sample_data/sample.txt","r") as file:
    print(file.tell())
    print(file.read(5))
    print(file.tell())
    print(file.read())
    file.seek(0)
    print(file.read())
    file.seek(6)
    print(file.tell())
    print(file.read())
```

## Practice 1
get the text from sample.txt and count the following
number of lines
number of words
number of characters
and save it to the counter_of_string.txt file

```python
from functools import reduce

with open("./sample_data/sample.txt","r") as file:
    strings=file.readlines()
    total_lines= len(strings)
print(strings)
print(total_lines)

# Total words
total_words = sum(len(line.split()) for line in strings)
print(total_words)
#using lambda
number_of_words= list(map(lambda x : len(x.split()), strings))
print(number_of_words)
total= reduce(lambda x,y: x+y, number_of_words)
print(total)

# Total characters
total_characters = sum(len(line)for line in strings)
print(total_characters)

with open("./sample_data/counter_of_string.txt","w") as file2:
    file2.write(f"Total Lines: {total_lines}\n")
    file2.write(f"Total Words; {total_words}\n")
    file2.write(f"Total Characters: {total_characters}\n")
    
with open("./sample_data/counter_of_string.txt","r") as file3:
    print(file3.read())
```
