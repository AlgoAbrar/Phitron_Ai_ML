## Tokenize

### splitting joining and formatted string

```python
prompt = "what is your name ?"

token = prompt.split()
print(token)

#join to convert into string again
new_string = " ".join(token)
print(new_string)
print(type(new_string))
```

```python
#formatting string
name = "Abrar"
age=24
height= 6.01123445

#print(f"His name is {name} and his age is {age} and his height is {height:.2f} feet")
info= (f"His name is {name} and his age is {age} and his height is {height:.2f} feet")

print(info)
print(type(info))


model_accuracy= 0.83333
print(f"The model accuracy is {model_accuracy:.2%}")
```

## List

```python
#List is a collection of Data

numbers=[100,20,30,40,50]
print(type(numbers))
```

```python
# accessing the element
print(numbers[0])
```

```python
numbers[2]=90

print(numbers[2])
```

```python
#slicing

new_list=numbers[0:3]
print(new_list)
```

```python
#modification

numbers.append(89)

print(numbers)
```

```python
numbers.insert(4,47) #at any position
print(numbers)
```

```python
#deleting

numbers.pop() #last element delete kore
print(numbers)
```

```python
# List in List

list_in_list = [1,2,3,[4,5,6],7,8]
print(type(list_in_list))
```
