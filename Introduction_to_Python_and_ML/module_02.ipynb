## Conditional Statement

```python
#if, else, elif
raining= True

if raining:
    print("Take an umbrella") #indentation is important in python with 4 spaces
    print("Walk with caution")
else:
    print("Enjoy the weather")
```

```python
raining= False
muddy= True
#nested if-else

if raining:
    print("Take an umbrella") #indentation is important in python with 4 spaces
    print("Walk with caution")
    
    if muddy:
        print("Wear boots") 
    else:
        print("Wear shoes")

else:
    print("Enjoy the weather")
```

```python
taka = 370

if taka > 500 and taka <= 1000:
    print("I will go out")
elif taka > 400 and taka <= 500:
    print("I will order pizza")
elif taka > 300 and taka <= 400:
    print("I will order biryani")
else:
    print("I will eat at home")
```

```python
#more than 1 condition
taka=1000
raining= True

if taka > 500 and raining:
    print("I will not go out")
    print("I will order food")
elif taka > 500 and not raining:
    print("I will go out")
elif taka <= 500 and raining:
    print("I will order food")
elif taka <= 500 and not raining:
    print("I will eat at home")
else:
    print("I am broke")
```

## Loop in Python

```python
for i in range(10):
    print(i, "Hello World")

list(range(10))
```

```python
#range can have 3 parameters, range(start, stop, step)
#start by default is 0, (stop is required), step by default is 1

for i in range(0, 10, 2):
    print(i, "Hello World")
```

```python
for i in range(1, 11):
    for j in range(1, 11):
        print(i, j)
```

## While Loop

```python
# until a specific condition is met, we can use while loop

# while loop
# while condition:
#     # code to be executed
i = 0
while i < 10:
    print(i, "Hello World")
    i += 1 # i = i + 1
```
