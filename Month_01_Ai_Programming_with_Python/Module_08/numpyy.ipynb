# Numerical Python

```python
import numpy as np

# Ndarray

arr1=np.array([1,2,3,4,5])

print(type(arr1))
print(f"Dimension of array is {arr1.ndim}")

# 2d Array

arr2= np.array([[1,2,3,], # row 1
                [4,5,6]]) # row 2

print(f"Dimension of array is {arr2.ndim}")


# 3d array

arr3= np.array([
               #1st floor
               [ [1,2,3], # row 1
                [4,5,6]] ,  # row 2
               # 2nd floo
               
               [ [1,2,3,], # row 1
                [4,5,6]], # row 2
               
               #3rd floor
               [ [1,2,3,], # row 1
                [4,5,6]]] # row 2 
               )

print(f"Dimension of array is {arr3.ndim}")
```

## Attribute of Array

```python
# dimension
print("Dimensions")
print(arr1.ndim)
print(arr2.ndim)
print(arr3.ndim)

#shape
print("")
print("Shape")
print(arr1.shape)
print(arr2.shape)
print(arr3.shape)

# Data type
print("")
print("Data Type")
print(arr1.dtype) # row 5 col 0
print(arr2.dtype) # row 2 column 3
print(arr3.dtype) # floor 3 row 2 col 3

#Size
print("")
print("Size")
print(arr1.size)
print(arr2.size)
print(arr3.size)

```

```python
# dictionary

dc = { 'a' : 10 , 'b' : 20 , 'c' :30 }

keys = dc.keys()
values = dc.values()
items = dc.items()
```

```python
#keys to ndarray
arr = np.array(list(keys))
print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
print(arr.size)
```

```python
#values to ndarray
arr = np.array(list(values))
print(arr)
arr = arr.astype(np.int16)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
print(arr.size)
```

```python
#items to ndarray
arr = np.array(list(items))
print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
print(arr.size)
```

# array creation with random values

# array creation with range functions

```python
# np.arange(first , end , step ) -> similar to range function

arr = np.arange(1 , 10 ,1 )
mat = arr.reshape(3,3)

print(arr)
print(mat)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
print(arr.size)
```

```python
# np.linspace(start , end(included) , koita)

arr = np.linspace(0, 4 ,6)


print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
print(arr.size)
```

```python
# np.logspace(start , end(included) , koita )

arr = np.logspace(0 , 4 ,6, base =2)

print(arr)
```

# Indexing and Slicing

```python
#indexing

print(arr1)
print(arr2)

# arr1[2] = 100

# print(arr1)
arr2[0][2]= 100

print(arr2)
```

```python
#slicing
#1d = arr[start:end:step]

arr1_mod = arr1[1:4].copy()
arr1[1] = 2


arr1_mod[2] = 200
print(arr1_mod)
print(arr1)
```

```python
#2d slicing
# arr[ row_start : row_end : step , column_start : column_end : step ]

print(arr2)
print("")
#getting a row

row_0 = arr2[0:1 ,]

row_1= arr2[1: , ]

print(row_0)
print(row_1)
print("")

#getting a column

col_0 = arr2[:: ,0:1]
col_1 = arr2[::,1:2]

print(col_0)
print("")
print(col_1)

#getting a portion


portion = arr2[:: , 0:2]
print("")
print(portion)
```
