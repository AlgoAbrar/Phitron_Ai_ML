# Object Oriented Programming with Python

### Attributes are variable name, Method is function() name

Class and Object

example: Phone
There can be multiple phone brands, type, battery, processor, camera etc.
Everything has some common structure

Object is call instance in other language

```python
# base class / parent class
class Phone:
  category = "Electronics"

  # constructor
  def __init__(self,model , battery , camera,battery_percentage=100):
    self.model = model
    self.battery = battery
    self.camera = camera
    self.battery_percentage = battery_percentage

  # methods
  def charge(self,hour):
    print(f"charge completed by {hour} for {self.model}")

  def capture(self,photo):
    if self.battery_percentage <= 0:
      print("No charge")
    else:
      self.battery_percentage-=photo
      print(f"photo captured in {self.model}")



class Cooling_Mechanism:
  def __init__(self,cooling_method):
    self.cooling_method= cooling_method


  #methods
  def cooling_on(self):
    print(f"the system is being cool by {self.cooling_method}")
```

```python
# Creating object

apple = Phone("Iphone17",3000,40)
blueberry = Phone("B-17",4000,30)
motorola = Phone("M-17",3500,35)
apple.capture(10)
print(motorola.battery_percentage)

# no values
```

```python
# Attribute dite hobe
apple= Phone("Iphone 17", 3000, 40)
blueberry = Phone("B-17", 4000, 12)
motorola = Phone("G50", 3500, 33)
```

```python
print(apple.camera)
print(motorola.camera)
print(motorola.battery)
```

# Inheritance

```python
# Single Inheritance

# child / derived class
class SmartPhone(Phone):
  def __init__(self,model , battery , camera,processor):
    super().__init__(model,battery,camera)
    self.processor = processor

   ## modified method
  def charge(self,hour):
    print("fast charging in process")
    super().charge(hour)
```

```python
pro = SmartPhone("X",5000,100,"SnapDragon")

print(pro.model)

pro.charge(1)
```

## Multiple Inheritance

```python
class SmartPhone_CoolingMode(SmartPhone,Cooling_Mechanism):
   def __init__(self,model , battery , camera,processor,cooling_method):
    SmartPhone.__init__(self,model,battery,camera,processor) # jehetu multiple korci
    Cooling_Mechanism.__init__(self,cooling_method)
```

```python
pro_cooling = SmartPhone_CoolingMode("Y",5000,100,"SD","Nitrogen")

print(pro_cooling.processor) # SmartPhone class theke
print(pro_cooling.battery) # Phone class theke
print(pro_cooling.cooling_method) # Cooling_Mechanism class theke
print(pro_cooling.cooling_on()) # Cooling_Mechanism class er method
print(pro_cooling.charge(1)) # SmartPhone er modified charge jeta Phone class theke inherited
```

# Polymorphism

```python
# Polymorphism mean Many Form
# Method Overriding for various form

class Camera:
    def __init__(self, name):
        self.name= name
    # method
    def capture(self):
        print("A photo is capture")
```

```python
class Smart(Camera):
    def __init__(self, name, resolution):
        super().__init__(name)
        self.resolution = resolution
    #overriding 
    def capture(self):
        print("Photo is captured by phone")
        
class DSLR(Camera):
    def __init__(self, name, resolution):
        super().__init__(name)
        self.resolution = resolution
    def capture(self):
        print("Photo is captured by DSLR")
        
class Drone(Camera):
    def __init__(self, name, resolution):
        super().__init__(name)
        self.resolution = resolution
    
    def capture(self):
        print("Photo is captured by drone")
        
        
phone = Smart("phone", 30)
dslr= DSLR ("DSLR", 200)
drone= Drone("Drone", 150)

phone.capture()
dslr.capture()
drone.capture()
```

# Encapsulation

```python
# It capsulate Data and Method for security

class Mobile:
    def __init__(self, name, model, imei):
        self.name= name
        self.model= model
        self.imei= imei
        
    def charge(self):
        print("Phone is charging")
        
iphone= Mobile("Phone", '17', '1xkaf1')

print(iphone.name)
print(iphone.model)
print(iphone.imei)
iphone.imei="123"
iphone.name= "Abrar"

print(iphone.name)
print(iphone.imei)
```

```python
class Mobile:
    def __init__(self, name, model, imei):
        self.name= name
        self.model= model
        self.__imei= imei  # double underscore means private
        
    def charge(self):
        print("Phone is charging")
        
iphone= Mobile("Phone", '17', '1xkaf1')

print(iphone.name)
print(iphone.model)
print(iphone.imei) # it is private so error will show

iphone.name= "Abrar"

print(iphone.name)
```

```python
# Designer Part

class Mobile:
    def __init__(self, name, model, imei):
        self.__name= name
        self.__model= model
        self.__imei= imei  # double underscore means private
        
    def charge(self):
        print("Phone is charging")
        
    def imei_getter(self): # we will get Imei because of this method only
        return self.__imei
    
    def model_getter(self):
        return self.__model
    
    def name_getter(self):
        return self.__name
    
    # what if user want to change name
    def name_setter(self, name):
        self.__name=name



# Outside Part
     
iphone= Mobile("Phone", '17', '1xkaf1')

print(iphone.name)
print(iphone.model)
print(iphone.imei_getter()) # it will show because of the function has access to private and function is public

iphone.name= "Abrar"
iphone.imei= "123"
print(iphone.name)
print(iphone.imei_getter())
```

```python
# Designer Part

class Mobile:
    def __init__(self, name, model, imei):
        self.__name= name
        self.__model= model
        self.__imei= imei  # double underscore means private
        
    def charge(self):
        print("Phone is charging")
        
    def imei_getter(self): # we will get Imei because of this method only
        return self.__imei
    
    def model_getter(self):
        return self.__model
    
    def name_getter(self):
        return self.__name
    
    # what if user want to change name
    def name_setter(self, name):
        self.__name=name



# Outside Part
     
iphone= Mobile("Phone", '17', '1xkaf1')

print(iphone.name_getter())
print(iphone.model_getter())
print(iphone.imei_getter()) # it will show because of the function has access to private and function is public

iphone.name_setter("Samsung")
iphone.imei= "123"
print(iphone.name_getter())
print(iphone.imei_getter())
```
